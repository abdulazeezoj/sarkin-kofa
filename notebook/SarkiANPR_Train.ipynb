```python
import ultralytics
from roboflow import Roboflow
from ultralytics import YOLO
import shutil
import os


ultralytics.checks()
```

```python
HOME: str = os.getcwd()
print(HOME)
```

```python
# Set up the Roboflow Dataset Directory
!mkdir {HOME}/datasets
%cd {HOME}/datasets

# Download the dataset
rf = Roboflow(api_key="sOd6IHKIcXdroCfAXoWn")
project = rf.workspace("abdulazeez-jimoh-1qumc").project("sarkin-kofa-anpr")
dataset = project.version(2).download("yolov8")
```

```python
# Change to the home directory
%cd {HOME}

# Load the model
name = "anpr"
resume = False
yolov8n = YOLO(f"{name}-last.pt") if resume else YOLO("yolov8n.pt")
epochs = 50 if resume else 100

# Train the model
yolov8n.train(
    data=f"{dataset.location}/data.yaml", epochs=epochs, imgsz=640, name=name, resume=resume
)

# Export the model
yolov8n.export(format="onnx", imgsz=640, half=True, simplify=True)

# Copy the model
shutil.copy2(f"runs/detect/{name}/weights/best.onnx", f"{name}.onnx")
shutil.copy2(f"runs/detect/{name}/weights/best.pt", f"{name}.pt")
shutil.copy2(f"runs/detect/{name}/weights/last.pt", f"{name}-last.pt")
```
